## Week 3-3：BeautifulSoup HTML 解析

### 核心觀點
BeautifulSoup 將複雜的 HTML 轉換為易於操作的 Python 物件。

### 程式碼




````python
from bs4 import BeautifulSoup
html_content = response.text
soup = BeautifulSoup(html,"html.parser")
# 嘗試尋找 "div" 標籤與特定類別，這個類別名稱看起來像是一個自動生成的ID，可能不穩定或不存在。
# 如果找不到匹配的元素，soup.find() 會返回 None。

# 1. Extract the page title
page_title = soup.find("title").text.strip()
print(f"網頁標題: {page_title}")

# 2. Extract options from the '計價幣別' dropdown
currency_dropdown = soup.find("select", id="ctl00_ContentPlaceHolder1_DD_classCurrent")

if currency_dropdown:
    print("\n計價幣別選項:")
    for option in currency_dropdown.find_all("option"):
        value = option.get("value")
        text = option.text.strip()
        print(f"  - {text} (值: {value})")
else:
    print("\n錯誤: 未找到計價幣別下拉選單 (ID: ctl00_ContentPlaceHolder1_DD_classCurrent)")
````

## Week 3-4：CSS 選擇器應用

### 核心觀點
CSS 選擇器是強大的元素定位工具，可精確定位 HTML 元素。

### 程式碼



````python
# ID選擇器: 尋找 id="aspnetForm" 的表單
form_element = soup.select_one("#aspnetForm")
print("--- ID 選擇器 --- ")
print(f"找到 ID 為 'aspnetForm' 的表單: {form_element.name if form_element else '未找到'}")

# Class選擇器: 尋找所有 class="group_item" 的元素
group_items = soup.select(".group_item")
print("\n--- Class 選擇器 --- ")
print(f"找到 {len(group_items)} 個 class 為 'group_item' 的元素。")
if group_items:
    print("前三個元素的標籤名稱:")
    for i, item in enumerate(group_items[:3]):
        print(f"  - {item.name}")

# 屬性選擇器: 尋找所有 href 屬性以 'http://fund.cnyes.com/' 開頭的 <a> 標籤
fund_links = soup.select("a[href^='http://fund.cnyes.com/']")
print("\n--- 屬性選擇器 --- ")
print(f"找到 {len(fund_links)} 個 href 以 'http://fund.cnyes.com/' 開頭的連結。")
if fund_links:
    print("前三個連結的 href 和文字:")
    for i, link in enumerate(fund_links[:3]):
        print(f"  - href: {link.get('href')}, Text: {link.text.strip()}")
````

## Week 3-5：實戰 - PTT 爬蟲

### 核心觀點
透過實際案例學習如何爬取公開網頁資料。

### 程式碼



````python
import requests
from bs4 import BeautifulSoup
import time
import random

# User-Agent
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"}
url = "https://www.ptt.cc/bbs/Tech_Job/index.html"

# 添加隨機延遲，模擬人類行為，避免被偵測為爬蟲
time.sleep(random.uniform(2, 5)) # 延遲 2 到 5 秒之間的隨機時間

try:
    response = requests.get(url, headers=headers)
    response.raise_for_status()  # 檢查請求是否成功 (200 OK)
    soup = BeautifulSoup(response.text,"html.parser")
    print("Successfully fetched PTT page.")
    # 您可以在這裡繼續您的解析邏輯
    # 例如：找到所有文章標題
    titles = soup.find_all('div', class_='title')
    for t in titles:
        if t.a: print(t.a.text.strip())
except requests.exceptions.RequestException as e:
    print(f"Request failed: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
````

## Week 3-6：實戰 - PCHome爬蟲

### 核心觀點
電商網站資料是商業分析的重要來源。

### 程式碼




````python
import requests
import time
import random

def scrape_pchome(keyword):
    # PChome 的搜尋 API 介面
    url = f"https://ecshweb.pchome.com.tw/search/v3.3/all/results?q={keyword}&page=1&sort=rnk/dc"

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    }

    print(f"正在透過 PChome 搜尋: {keyword}...")

    # 模擬人類行為
    time.sleep(random.uniform(2, 4))

    try:
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()

        data = response.json()
        items = data.get('prods', [])

        if not items:
            print("未找到任何商品。")
            return

        print(f"成功獲取 {len(items)} 筆商品資料：\n")

        for item in items[:10]:  # 列出前 10 筆
            name = item.get('name')
            price = item.get('price')
            # 組合商品連結
            prod_id = item.get('Id')
            link = f"https://24h.pchome.com.tw/prod/{prod_id}"

            print(f"品名: {name}")
            print(f"價格: ${price}")
            print(f"連結: {link}")
            print("-" * 30)

    except requests.exceptions.ConnectionError:
        print("連線失敗：此環境可能屏蔽了電商網域，請嘗試在『本地電腦』執行此程式。")
    except Exception as e:
        print(f"發生錯誤: {e}")

# 執行測試
scrape_pchome("DDR5")
````

## Week 3-9：反爬蟲機制與應對

### 核心觀點
了解網站的反爬蟲機制，學習合法的繞過技巧。

### 程式碼




````python
import requests
import time
import random

# 核心觀點：了解網站的反爬蟲機制，學習合法的繞過技巧。

# 1. User-Agent：模擬瀏覽器行為，避免被偵測為爬蟲。
# 使用一個常見的瀏覽器 User-Agent 字串
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"}

# 2. 代理IP：隱藏真實 IP，避免 IP 被封鎖。
# 這裡只是一個範例，您需要替換成實際可用的代理 IP
# 例如：proxies = {"http": "http://your_proxy_ip:port", "https": "https://your_proxy_ip:port"}
proxies = None # 預設為 None，如果您有代理IP請在此設定

# 3. 延遲：模擬人類瀏覽行為，避免請求過於頻繁。
# 添加隨機延遲，模擬人類行為，避免被偵測為爬蟲。
min_delay = 1  # 最小延遲秒數
max_delay = 3  # 最大延遲秒數

# 範例網址 (請替換為您想要爬取的實際網址)
example_url = "https://httpbin.org/get" # 測試用，會回傳您的請求資訊
# example_url = "https://www.google.com" # 更常見的測試網址

print(f"正在對 {example_url} 發送請求...")
time.sleep(random.uniform(min_delay, max_delay)) # 執行隨機延遲

try:
    # 發送 HTTP GET 請求
    response = requests.get(example_url, headers=headers, proxies=proxies, timeout=10)
    response.raise_for_status() # 檢查請求是否成功 (200 OK)

    print(f"請求成功！狀態碼: {response.status_code}")
    print("部分回應內容 (前500字元):\n", response.text[:500])

    if proxies:
        print("使用的代理IP已設定完成。")
    else:
        print("未使用代理IP，如果您需要隱藏真實IP，請設定 'proxies' 變數。")

except requests.exceptions.RequestException as e:
    print(f"請求失敗: {e}")
    if "403 Client Error" in str(e):
        print("可能是網站偵測到爬蟲，請檢查 User-Agent、延遲或考慮使用代理IP。")
    elif "Connection refused" in str(e) and proxies:
        print("代理IP可能無法連接或無效。")
except Exception as e:
    print(f"發生未知錯誤: {e}")
````

## Week 9-2：Pandas Merge 與 Concat

### 核心觀點
資料合併是整合多個資料源的必要技能。

### 程式碼



````python
import pandas as pd

# ==========================================
# 1. 先建立示範用的 df1 與 df2
# ==========================================
data1 = {
    "ID": ["TXN001", "TXN002", "TXN003"],
    "Department": ["Sales", "Marketing", "IT"],
    "Name": ["Alice", "Bob", "Charlie"]
}
df1 = pd.DataFrame(data1)

data2 = {
    "ID": ["TXN002", "TXN003", "TXN004"],
    "Salary": [50000, 85000, 45000],
    "Experience": [3, 8, 2]
}
df2 = pd.DataFrame(data2)

print("=== 原始資料表 ===")
print("df1:\n", df1)
print("\ndf2:\n", df2)
print("-" * 30)

# ==========================================
# 2. 執行你的 Week 9-2 程式碼
# ==========================================

# Concat - 串聯 (上下拼接，axis=0)
# 加入 ignore_index=True 讓索引重新排序會比較整齊
result_concat = pd.concat([df1, df2], axis=0, ignore_index=True)
print("\n=== Concat 串聯 ===")
print(result_concat)

# Merge - 關聯式合併 (預設 how="inner"，只保留兩邊都有的 ID)
result_merge_inner = pd.merge(df1, df2, on="ID")
print("\n=== Merge (Inner) ===")
print(result_merge_inner)

# Merge - 關聯式合併 (how="left"，以左邊 df1 的 ID 為主)
result_merge_left = pd.merge(df1, df2, on="ID", how="left")
print("\n=== Merge (Left) ===")
print(result_merge_left)

# Merge - 關聯式合併 (how="outer"，保留兩邊所有的 ID)
result_merge_outer = pd.merge(df1, df2, on="ID", how="outer")
print("\n=== Merge (Outer) ===")
print(result_merge_outer)

# join - 主要是基於 Index (索引) 來合併
# 小提醒：Pandas 的 join 預設是看 Index，所以我們需要先把 ID 設為 Index 再 join
result_join = df1.set_index("ID").join(df2.set_index("ID"), how="inner")
print("\n=== Join (Inner) ===")
print(result_join)
````

## Week 14-2：KNN 分類器

### 核心觀點
KNN 是基於實例的學習，透過最近鄰居進行分類。

### 程式碼



````python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris

iris = load_iris()
X,y = iris.data,iris.target

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

# 找最佳K
k_values = range(1,21)
scores = []
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    scores.append(knn.score(X_test,y_test))

best_k = k_values[np.argmax(scores)]
print(f"最佳K:{best_k},準確率:{max(scores):.4f}")
````

## Week 15-3：過擬合與欠擬合處理

### 核心觀點
理解模型複雜度與泛化能力的平衡。

### 程式碼


````python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris # Import load_iris
import matplotlib.pyplot as plt # Import matplotlib for plotting

# 載入資料 (確保 X 和 y 來源一致，這裡使用 iris dataset)
iris = load_iris()
X, y = iris.data, iris.target

# 測試不同深度
depths = range(1,20)
train_scores,test_scores = [],[]

for depth in depths:
    dt = DecisionTreeClassifier(max_depth=depth,random_state=42)
    dt.fit(X,y)
    train_scores.append(dt.score(X,y))
    cv_scores = cross_val_score(dt,X,y,cv=5)
    test_scores.append(cv_scores.mean())

# 視覺化
plt.plot(depths,train_scores,"b-o",label="Train")
plt.plot(depths,test_scores,"r-o",label="CV")
plt.xlabel("Max Depth") # 添加X軸標籤
plt.ylabel("Accuracy Score") # 添加Y軸標籤
plt.title("Train vs Cross-validation Accuracy by Max Depth") # 添加標題
plt.legend()
plt.show() # 顯示圖表

# 解決過擬合: 降低複雜度、正則化、增加資料
# 解決欠擬合: 增加複雜度、增加特徵
````
